```python
'14장 딥러닝'
# Keras를 이용한 MNIST 숫자 인식

import matplotlib.pyplot as plt
import tensorflow as tf

# mnist 데이터셋
mnist = tf.keras.datasets.mnist

# train/test 데이터 분리
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# 0~1사이로 정규화
X_train, X_test = X_train/255.0, X_test/255.0

# 모델
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten(input_shape=(28, 28)))      # 28x28을 flatten해서 한 열 짜리 벡터로 -> 784
model.add(tf.keras.layers.Dense(512, activation='relu'))
model.add(tf.keras.layers.Dropout(0.2))               # 드롭아웃 20퍼 -> 과대적합 방지
model.add(tf.keras.layers.Dense(10, activation='softmax'))    # softmax 활성화 함수: 다중 클래스 분류에 사용

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=5, verbose=0)

model.summary()

score = model.evaluate(X_test, y_test, verbose=0)
print("테스트 loss값: ", score[0])
print("테스트 accuracy: ", score[1])
```

```python
#### 2. CNN 이용
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import datasets, layers, models

fashion_mnist = keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()

train_images = train_images.reshape((60000, 28, 28, 1))
test_images = test_images.reshape((10000, 28, 28, 1))

train_images = train_images/255.0
test_images = test_images/255.0

##### 모델
## CNN은 '입력'을 flatten하지 않음 -> 2차원 입력
model = models.Sequential()

# 1) 컨볼루션-풀링 
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))

# 2) 다층신경망
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model.fit(train_images, train_labels, epochs=5)

test_loss, test_acc = model.evaluate(test_images, test_labels)
```
