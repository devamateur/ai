```python
a =  np.array([[1, 1], [0, 1]])
b = np.array([[2, 0], [3, 4]])

# 곱하기와 dot 연산 구분
# dot: 두 벡터의 내적
print(a*10)   # [10, 10], [0, 10]
print(a+b)    # [3, 1], [3, 5]
print(a.dot(b))   # a와 b의 내적  [2+3, 0+4], [0+3, 0+4]
```

```python
'선형회귀 - 경사하강법'
# 1. 하나하나 직접 
import numpy as np
import matplotlib.pyplot as plt

X = np.array([0.0, 1.0, 2.0])   # [1.0, 2.0, 3.0]
y = np.array([3.0, 3.5, 5.5])   # [2.0, 5.0, 6.0]

W = 0   # 기울기
b = 0   # 절편

lrate = 0.01   # 학습률
epochs = 200 # 반복 횟수

loss = []

n = float(len(X))   # 입력 데이터의 개수

# 경사하강법
for i in range(epochs):
  y_pred = W*X + b    # 선형회귀식
  loss.append(round(np.sqrt(np.sum(np.square(y-y_pred))), 2))
  print(i, ": ", loss[i])

  # loss값이 변하지 않을 경우 break
  if i != 0 and loss[i] == loss[i-1]:
    print("loss값이 기존과 같습니다!")
    print("이전 epoch의 loss: ", loss[i-1], "현재 loss: ", loss[i])
    break

  dW = (2/n) * sum(X * (y_pred-y))
  db = (2/n) * sum(y_pred - y)
  W = W-lrate * dW    # 기울기 수정
  b = b-lrate * db    # 절편 수정

# 기울기와 절편
print (W, b)
# y_pred
y_pred = W*X + b

# 시각화
plt.scatter(X, y)

plt.plot([min(X), max(X)], [min(y_pred), max(y_pred)], color='red')
plt.show()

```

```python

# 2. 파이썬 라이브러리 이용
import matplotlib.pylab as plt
from sklearn import linear_model

lr = linear_model.LinearRegression()

X = [[0], [1], [2]]
y = [3, 3.5, 5.5]

# fit
lr.fit(X, y)

plt.scatter(X, y, color='black')

# predict
y_pred = lr.predict(X)

plt.plot(X, y_pred, color='blue', linewidth=3)
plt.show()
```

```python
import matplotlib.pylab as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import datasets

# 당뇨병 데이터
diabetes = datasets.load_diabetes()

# train/test 분리
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(diabetes.data, diabetes.target, test_size=0.2, random_state=0)

# 모델 fit
model = LinearRegression()
model.fit(X_train, y_train)

# predict
y_pred = model.predict(X_test)

# 실제 데이터와 pred 비교
plt.plot(y_test, y_pred, '.')

x = np.linspace(0, 330, 100)
y = x
plt.plot(x, y)
plt.show()
```
