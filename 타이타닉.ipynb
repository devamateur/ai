```python
train['Embarked'].value_counts()      # Embarked 컬럼의 내용 확인(S가 가장 많다)
```

```python
train["Embarked"] = train["Embarked"].fillna("S")     # 결측치를 가장 많은 S로 대체
```

```python
# Embarked의 S, C, Q를 0, 1, 2로 매핑
train["Embarked"][train["Embarked"] == "S"] = 0
train["Embarked"][train["Embarked"] == "C"] = 1
train["Embarked"][train["Embarked"] == "Q"] = 2
```

```python
# test 매핑
test["Embarked"][test["Embarked"] == "S"] = 0
test["Embarked"][test["Embarked"] == "C"] = 1
test["Embarked"][test["Embarked"] == "Q"] = 2
```

```python
# Fare의 결측치는 평균값으로 대체
test["Fare"].fillna(df["Fare"].mean(), inplace=True)
```

```python
# 필요없는 컬럼 제거하기
train = train.drop(['Name'], axis=1)
test = test.drop(['Name'], axis=1)
```

```python
train = train.drop(['PassengerId'], axis=1)
test = test.drop(['PassengerId'], axis=1)
```

```python
train = train.drop(['Cabin'], axis=1)
test = test.drop(['Cabin'], axis=1)
```

```python
train = train.drop(['Ticket'], axis=1)
test = test.drop(['Ticket'], axis=1)
```

```python
# 성별 매핑
sex_mapping = {"male": 0, "female": 1}
train['Sex'] = train['Sex'].map(sex_mapping)
test['Sex'] = test['Sex'].map(sex_mapping)
```

```python
from sklearn.model_selection import train_test_split

# 예측해야 할 Survived를 따로 떼어냄
pred = train.drop(['Survived'], axis=1)
target = train["Survived"]

x_train, x_test, y_train, y_test = train_test_split(pred, target, test_size=0.3)
```

```python
# 로지스틱 회귀

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

logr = LogisticRegression()
logr.fit(x_train, y_train)

y_pred = logr.predict(x_test)
acc_logr = round(accuracy_score(y_pred, y_test)*100, 2)
print(acc_logr)
```

```python
# 의사결정나무
from sklearn.tree import DecisionTreeClassifier

decisiontree = DecisionTreeClassifier()
decisiontree.fit(x_train, y_train)
y_pred = decisiontree.predict(x_test)
acc_decisiontree = round(accuracy_score(y_pred, y_test)*100, 2)
print(acc_decisiontree)
```

```python
# 랜덤 포레스트
from sklearn.ensemble import RandomForestClassifier

randomforest = RandomForestClassifier()
randomforest.fit(x_train, y_train)
y_pred = randomforest.predict(x_test)
acc_randomforest = round(accuracy_score(y_pred, y_test)*100, 2)
print(acc_randomforest)
```

```python
# knn
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
acc_knn = round(accuracy_score(y_pred, y_test)*100, 2)
print(acc_knn)
```

```python
models = pd.DataFrame({
    'Model': ['KNN', 'Logistic Regression', 'Random Forest', 'Decision Tree'],
    'Score': [acc_knn, acc_logr, acc_randomforest, acc_decisiontree]
})
```

```python
# 각 모델들을 점수를 기준으로 내림차순 정렬
models.sort_values(by='Score', ascending = False)     
```
